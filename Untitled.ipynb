```python
import pandas as pd
import requests
```

```python
# Dallas coordinates
latitude  = 41.8781
longitude = -87.6298

start_date = "2024-01-01"
end_date   = "2024-05-18"

hourly_vars = [
    "temperature_2m",
    "precipitation",
    "windspeed_10m",
    "relative_humidity_2m"
]

url = "https://archive-api.open-meteo.com/v1/archive"  # or Historical endpoint
params = {
    "latitude": latitude,
    "longitude": longitude,
    "start_date": start_date,
    "end_date":   end_date,
    "hourly":     ",".join(hourly_vars),
    "temperature_unit": "celsius",
    "windspeed_unit": "kmh",
    "precipitation_unit": "mm",
    "timezone": "auto"
}

response = requests.get(url, params=params, timeout=120)
response.raise_for_status()

data = response.json()  
# Example: parse into DataFrame:
df = pd.DataFrame(data['hourly'])
df['time'] = pd.to_datetime(df['time'])

# Add metadata
df['Latitude']  = latitude
df['Longitude'] = longitude
df['Location']  = "Chicago"

print(df.head())
```

```python
# Make a copy to work with
real_chicago = df.copy()

# Split datetime into date/time components
real_chicago['Year']  = real_chicago['time'].dt.year
real_chicago['Month'] = real_chicago['time'].dt.month
real_chicago['Day']   = real_chicago['time'].dt.day
real_chicago['Time']  = real_chicago['time'].dt.time

# Map to synthetic-style column names
real_chicago['Temperature_C']   = real_chicago['temperature_2m']
real_chicago['Precipitation_mm'] = real_chicago['precipitation']
real_chicago['Wind_Speed_kmh']   = real_chicago['windspeed_10m']
real_chicago['Humidity_pct']     = real_chicago['relative_humidity_2m']

# Keep only the columns matching your synthetic schema
real_chicago_aligned = real_chicago[[
    'Location',
    'Temperature_C',
    'Humidity_pct',
    'Precipitation_mm',
    'Wind_Speed_kmh',
    'Time',
    'Year',
    'Month',
    'Day'
]]

print(real_chicago_aligned.head())
print(real_chicago_aligned.shape)
```

```python
out_path = r"C:\PythonGIS\dataanalysis\proj\real\weather_Chicago_2024_JanMay.csv"

import os
os.makedirs(os.path.dirname(out_path), exist_ok=True)

real_chicago_aligned.to_csv(out_path, index=False)
print("Saved:", out_path)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Extract columns
precipitation = chicago_weather["Precipitation_mm"]
humidity = chicago_weather["Humidity_pct"]

# Calculate correlation
corr_value = chicago_weather[["Precipitation_mm", "Humidity_pct"]].corr().iloc[0,1]
print("Correlation (Precipitation vs Humidity):", corr_value)

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(precipitation, humidity, alpha=0.3, s=10)
plt.xlabel("Precipitation, mm")
plt.ylabel("Humidity, %")
plt.title(f"Chicago Real Data: Precipitation vs Humidity\nCorrelation = {corr_value:.3f}")

# Add trendline
slope, intercept, r_value, p_value, std_err = linregress(precipitation, humidity)
plt.plot(precipitation, slope*precipitation + intercept, lw=2)

plt.show()
```
