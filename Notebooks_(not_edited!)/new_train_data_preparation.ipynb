```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
# to check weather patterns, firstly, I arrange bivariate correlation analysis between some parameters

# here is temperature and wind speed
```

```python
# Extract columns
temperature = montreal["Temperature_C"]
wind_speed = montreal["Wind_Speed_kmh"]

# Calculate correlation
corr_value = montreal[["Temperature_C", "Wind_Speed_kmh"]].corr().iloc[0,1]
print("Correlation (Temperature vs Wind Speed):", corr_value)

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(temperature, wind_speed, alpha=0.3, s=10)
plt.xlabel("Temperature (°C)")
plt.ylabel("Wind Speed (km/h)")
plt.title(f"Historical weather train data: Wind Speed vs Temperature\nCorrelation = {corr_value:.3f}")

# Add trendline
slope, intercept, r_value, p_value, std_err = linregress(temperature, wind_speed)
plt.plot(temperature, slope*temperature + intercept, lw=2)

plt.show()
```

```python
# no correlation between temperature and wind speed. It seems logic because temperature and wind speed usually independent parameters
```

```python
# in real world high pressure zone actually have dry and clean weather with lower humidity, while low pressure zone 
# actually brings clouds and precipitation

# lets check this pattern in next scatter plot
```

```python
# Extract columns
pressure = montreal["Pressure_hPa"]
humidity = montreal["Humidity_pct"]

# Calculate correlation
corr_value = montreal[["Pressure_hPa", "Humidity_pct"]].corr().iloc[0,1]
print("Correlation (Pressure vs Humidity):", corr_value)

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(pressure, humidity, alpha=0.3, s=10)
plt.xlabel("Pressure, hPa")
plt.ylabel("Humidity, %")
plt.title(f"Historical weather train data: Pressure vs Humidity\nCorrelation = {corr_value:.3f}")

# Add trendline
slope, intercept, r_value, p_value, std_err = linregress(pressure, humidity)
plt.plot(pressure, slope*pressure + intercept, lw=2)

plt.show()
```

```python
# here it can be seen weak negative correlation between pressure and humidity. It is not clean "cast-iron" pattern that
# in real conditions always low humidity in high pressure zone, but weak negative correlation seem logical here
```

```python
# Extract columns
pressure = montreal["Pressure_hPa"]
temperature = montreal["Temperature_C"]

# Calculate correlation
corr_value = montreal[["Pressure_hPa", "Temperature_C"]].corr().iloc[0,1]
print("Correlation (Pressure vs Temperature):", corr_value)

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(pressure, temperature, alpha=0.3, s=10)
plt.xlabel("Pressure, hPa")
plt.ylabel("Temperature, (°C)")
plt.title(f"Historical weather train data: Pressure vs Temperature\nCorrelation = {corr_value:.3f}")

# Add trendline
slope, intercept, r_value, p_value, std_err = linregress(pressure, temperature)
plt.plot(pressure, slope*pressure + intercept, lw=2)

plt.show()
```

```python
# weak negative correlation between pressure and temperature. It is logic, because, in higher latutude > 50 N, in higher pressure zone 
# actually lower temperature at ninght and morning than in low pressure zone with clouds and high humidity
```

```python
# in next step check diurnal cycle analysis of temperatue in Toronto
```

```python
# Convert Time column into datetime.time (if it's string)
toronto['Time'] = pd.to_datetime(toronto['Time'], format="%H:%M:%S").dt.time
```

```python
# create function that change time range into categorical format, part of day
def get_part_of_day(t):
    hour = t.hour
    
    if 1 <= hour <= 6:
        return "Night"
    elif 7 <= hour <= 12:
        return "Morning"
    elif 13 <= hour <= 18:
        return "Day"
    else:
        # Evening includes 19-23 and 00
        return "Evening"

# Create new column
toronto['Part_of_Day'] = toronto['Time'].apply(get_part_of_day)
toronto.head(10)
```

```python
# boxplot there range of temperature in each part of day
plt.figure(figsize=(8,5))
toronto.boxplot(column="Temperature_C", by="Part_of_Day")
plt.title("Temperature Distribution by Part of Day (Toronto historical)")
plt.suptitle("")
plt.xlabel("Part of Day")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
# calculate descriptive statistics for temperature
temp_by_part = toronto.groupby("Part_of_Day")["Temperature_C"].agg(["mean", "min", "max", "count"])
print(temp_by_part)
```

```python
# both boxplot and statistics output show logical diurnal cycle of temperature. Morning is coolost, evening is hottest.
# Actually it very depends of season, here annual analysis
```
